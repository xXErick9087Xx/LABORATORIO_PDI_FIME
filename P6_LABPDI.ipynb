```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# ==============================
# Cargar imágenes
# ==============================
print("Sube la primera imagen")
uploaded1 = files.upload()
img1_name = list(uploaded1.keys())[0]

print("Sube la segunda imagen")
uploaded2 = files.upload()
img2_name = list(uploaded2.keys())[0]

img1 = cv2.imread(img1_name)
img2 = cv2.imread(img2_name)

# ==============================
# Convertir a escala de grises
# ==============================
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# ==============================
# Ajustar tamaño si son diferentes
# ==============================
if img1_gray.shape != img2_gray.shape:
    print("Ajustando tamaño de la segunda imagen...")
    img2_gray = cv2.resize(img2_gray, (img1_gray.shape[1], img1_gray.shape[0]))

# ==============================
# Suma ponderada
# ==============================
alpha = 0.7  # peso imagen 1
beta = 0.3   # peso imagen 2

suma_ponderada = cv2.addWeighted(img1_gray, alpha, img2_gray, beta, 0)

# ==============================
# Suma promedio
# ==============================
suma_promedio = ((img1_gray.astype(np.float32) + img2_gray.astype(np.float32)) / 2).astype(np.uint8)

# ==============================
# Mostrar resultados
# ==============================
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(img1_gray, cmap='gray')
plt.title('Imagen 1')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(img2_gray, cmap='gray')
plt.title('Imagen 2')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(suma_ponderada, cmap='gray')
plt.title(f'Suma ponderada (α={alpha}, β={beta})')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(suma_promedio, cmap='gray')
plt.title('Suma promedio')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# ==============================
# Cargar imágenes
# ==============================
print("Sube la primera imagen")
uploaded1 = files.upload()
img1_name = list(uploaded1.keys())[0]

print("Sube la segunda imagen")
uploaded2 = files.upload()
img2_name = list(uploaded2.keys())[0]

img1 = cv2.imread(img1_name)
img2 = cv2.imread(img2_name)

# ==============================
# Convertir a escala de grises
# ==============================
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# ==============================
# Ajustar tamaño si es necesario
# ==============================
if img1_gray.shape != img2_gray.shape:
    print("Ajustando tamaño de la segunda imagen...")
    img2_gray = cv2.resize(img2_gray, (img1_gray.shape[1], img1_gray.shape[0]))

# ==============================
# Resta directa
# ==============================
resta_directa = img1_gray.astype(np.int16) - img2_gray.astype(np.int16)

# Normalizar para visualización (0-255)
resta_directa_norm = cv2.normalize(resta_directa, None, 0, 255, cv2.NORM_MINMAX)
resta_directa_norm = resta_directa_norm.astype(np.uint8)

# ==============================
# Resta con valor absoluto
# ==============================
resta_abs = cv2.absdiff(img1_gray, img2_gray)

# ==============================
# Mostrar resultados
# ==============================
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(img1_gray, cmap='gray')
plt.title('Imagen 1')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(img2_gray, cmap='gray')
plt.title('Imagen 2')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(resta_directa_norm, cmap='gray')
plt.title('Resta directa (normalizada)')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(resta_abs, cmap='gray')
plt.title('Resta absoluta (cv2.absdiff)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# ==============================
# Cargar imágenes
# ==============================
print("Sube la primera imagen")
uploaded1 = files.upload()
img1_name = list(uploaded1.keys())[0]

print("Sube la segunda imagen")
uploaded2 = files.upload()
img2_name = list(uploaded2.keys())[0]

img1 = cv2.imread(img1_name)
img2 = cv2.imread(img2_name)

# ==============================
# Convertir a escala de grises
# ==============================
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# ==============================
# Factores de multiplicación
# ==============================
factores = [0.5, 1.0, 1.5]


def aplicar_factor(img, factor):
    resultado = img.astype(np.float32) * factor
    resultado = np.clip(resultado, 0, 255)
    return resultado.astype(np.uint8)


img1_mod = [aplicar_factor(img1_gray, f) for f in factores]
img2_mod = [aplicar_factor(img2_gray, f) for f in factores]

# ==============================
# Mostrar resultados
# ==============================
plt.figure(figsize=(12,8))

# Imagen 1
plt.subplot(2,4,1)
plt.imshow(img1_gray, cmap='gray')
plt.title('Img1 Original')
plt.axis('off')

for i, f in enumerate(factores):
    plt.subplot(2,4,i+2)
    plt.imshow(img1_mod[i], cmap='gray')
    plt.title(f'Img1 x {f}')
    plt.axis('off')

# Imagen 2
plt.subplot(2,4,5)
plt.imshow(img2_gray, cmap='gray')
plt.title('Img2 Original')
plt.axis('off')

for i, f in enumerate(factores):
    plt.subplot(2,4,i+6)
    plt.imshow(img2_mod[i], cmap='gray')
    plt.title(f'Img2 x {f}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# ==============================
# Cargar imágenes
# ==============================
print("Sube la primera imagen")
uploaded1 = files.upload()
img1_name = list(uploaded1.keys())[0]

print("Sube la segunda imagen")
uploaded2 = files.upload()
img2_name = list(uploaded2.keys())[0]

img1 = cv2.imread(img1_name)
img2 = cv2.imread(img2_name)

# ==============================
# Convertir a escala de grises
# ==============================
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# ==============================
# Ajustar tamaño si es necesario
# ==============================
if img1_gray.shape != img2_gray.shape:
    print("Ajustando tamaño de la segunda imagen...")
    img2_gray = cv2.resize(img2_gray, (img1_gray.shape[1], img1_gray.shape[0]))

# ==============================
# Multiplicación píxel a píxel
# ==============================
# Convertir a float para evitar overflow
img1_float = img1_gray.astype(np.float32)
img2_float = img2_gray.astype(np.float32)

multiplicacion = img1_float * img2_float

multiplicacion_norm = cv2.normalize(multiplicacion, None, 0, 255, cv2.NORM_MINMAX)
multiplicacion_norm = multiplicacion_norm.astype(np.uint8)

# ==============================
# Mostrar resultados
# ==============================
plt.figure(figsize=(10,6))

plt.subplot(1,3,1)
plt.imshow(img1_gray, cmap='gray')
plt.title('Imagen 1')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(img2_gray, cmap='gray')
plt.title('Imagen 2')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(multiplicacion_norm, cmap='gray')
plt.title('Multiplicación píxel a píxel')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# ==============================
# Cargar imagen
# ==============================
print("Sube la imagen médica con bajo contraste")
uploaded = files.upload()
img_name = list(uploaded.keys())[0]

img = cv2.imread(img_name)

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# ==============================
# Histograma original
# ==============================
hist_original = cv2.calcHist([img_gray], [0], None, [256], [0,256])

# ==============================
# Ecualización de histograma
# ==============================
img_eq = cv2.equalizeHist(img_gray)

hist_eq = cv2.calcHist([img_eq], [0], None, [256], [0,256])

# ==============================
# Mostrar resultados
# ==============================
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(img_gray, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(2,2,2)
plt.plot(hist_original)
plt.title('Histograma Original')

plt.subplot(2,2,3)
plt.imshow(img_eq, cmap='gray')
plt.title('Imagen Ecualizada')
plt.axis('off')

plt.subplot(2,2,4)
plt.plot(hist_eq)
plt.title('Histograma Ecualizado')

plt.tight_layout()
plt.show()
```
