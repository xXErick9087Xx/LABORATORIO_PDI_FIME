```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files
import numpy as np

# Subir imagen desde la computadora
uploaded = files.upload()

# Obtener el nombre del archivo subido
nombre_imagen = list(uploaded.keys())[0]

# Leer la imagen
imagen_bgr = cv2.imread(nombre_imagen)

# Verificar si la imagen se cargó correctamente
if imagen_bgr is None:
    print("Error: no se pudo cargar la imagen.")
else:

    # Convertir de BGR a RGB para visualizar correctamente
    imagen_rgb = cv2.cvtColor(imagen_bgr, cv2.COLOR_BGR2RGB)

    # Convertir la imagen a HSV
    imagen_hsv = cv2.cvtColor(imagen_bgr, cv2.COLOR_BGR2HSV)

    # Separar canales H, S y V
    h, s, v = cv2.split(imagen_hsv)

    # Aplicar ecualización al canal de brillo
    v_ecualizado = cv2.equalizeHist(v)

    # Unir nuevamente los canales
    imagen_hsv_mejorada = cv2.merge((h, s, v_ecualizado))

    # Convertir de HSV a BGR
    resultado_bgr = cv2.cvtColor(imagen_hsv_mejorada, cv2.COLOR_HSV2BGR)

    # Convertir de BGR a RGB para mostrar
    resultado_rgb = cv2.cvtColor(resultado_bgr, cv2.COLOR_BGR2RGB)

    # Mostrar imágenes
    plt.figure(figsize=(12,6))

    # Imagen original
    plt.subplot(1,2,1)
    plt.imshow(imagen_rgb)
    plt.title("Imagen Original")
    plt.axis("off")

    # Imagen mejorada
    plt.subplot(1,2,2)
    plt.imshow(resultado_rgb)
    plt.title("Imagen Mejorada")
    plt.axis("off")

    plt.tight_layout()
    plt.show()
```
